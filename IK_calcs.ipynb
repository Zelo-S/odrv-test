```python
import math
L1 = 275.0 # mm
L2 = 275.0 # mm 
q1_max = 36 * math.pi / 180 # deg -> radians

print(math.pow(2 * L1 * math.cos(q1_max), 2))
print(4 * (math.pow(L1, 2) + math.pow(L2, 2)))
# xD_max_1 = (2 * L1 * math.cos(q1_max) + math.sqrt( math.pow(2 * L1 * math.cos(q1_max), 2) - 4 * (math.pow(L1, 2) + math.pow(L2, 2)))) / 2
# print(xD_max_1)
```

```python
def q1q2_from_xy(x, y):
    xD_mag = math.sqrt(x*x + y*y)
    c2 = (xD_mag * xD_mag - L1 * L1 - L2 * L2) / (2 * L1 * L2)
    if abs(c2) > 1:
        return None
    if c2 == 1:
        return math.atan2(y, x), 0, math.atan2(y, x), 0
    q2_1 = 1 * math.acos( c2 )
    q2_2 = -1 * math.acos( c2 )
    theta = math.atan2(y, x)
    q1_1 = theta - math.atan2(L2 * math.sin(q2_1), L1 + L2 * math.cos(q2_1))
    q1_2 = theta - math.atan2(L2 * math.sin(q2_2), L1 + L2 * math.cos(q2_2))
    return q1_1, q2_1, q1_2, q2_2

q1_1, q2_1, q1_2, q2_2 = q1q2_from_xy(275 + 275, 0)
print(q1_1, q2_1, q1_2, q2_2)

q1_1, q2_1, q1_2, q2_2 = q1q2_from_xy(450, 0)
print(math.degrees(q1_1), math.degrees(q2_1), math.degrees(q1_2), math.degrees(q2_2))
```
