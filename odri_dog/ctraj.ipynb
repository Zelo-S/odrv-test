```python
# Requires roboticstoolbox-python, ideally on python==3.10
%matplotlib ipympl
from spatialmath.base import *
from spatialmath import *
from roboticstoolbox import *
import numpy as np
import swift
from matplotlib import pyplot as plt
```

```python
mm = 0.001

L1 = -165 * mm
L2 = -165 * mm

leg2 = ERobot(ET.Ry() * ET.tz(L1) * ET.Ry() * ET.tz(L2))
leg2
```

```python
start_bottom_T = SE3.Trans(0, 0, -320 * mm)
end_top_T = SE3.Trans(0, 0, -200 * mm)
end2_bottom_T = SE3.Trans(0, 0, -320 * mm)
timesteps = np.arange(0, 1, 0.01)

def generate_traj(start: SE3, intermediate: SE3, end: SE3, timesteps: np.ndarray) -> np.ndarray:
    t = timesteps
    traj_up = ctraj(start, intermediate, t)
    res_up = leg2.ikine_LM(traj_up, mask=[1,1,1,1,0,1])

    traj_down = ctraj(intermediate, end, t)
    res_down = leg2.ikine_LM(traj_down, mask=[1,1,1,1,0,1])
    jointvals_up = res_up.q
    jointvals_down = res_down.q
    print(jointvals_up.shape)
    print(jointvals_down.shape)
    jointvals = np.concatenate((jointvals_up, jointvals_down), axis=0)
    print(jointvals.shape)
    return jointvals

generate_traj(start_bottom_T, end_top_T, end2_bottom_T, timesteps)
```
